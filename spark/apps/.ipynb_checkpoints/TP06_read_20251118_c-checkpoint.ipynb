## 1. text

https://spark.apache.org/docs/latest/sql-data-sources-text.html

https://spark.apache.org/docs/latest/api/python/reference/pyspark.sql/api/pyspark.sql.DataFrameReader.text.html#pyspark.sql.DataFrameReader.text

- Read a file and create a dataframe
    - default separator
    - choose "." as separator

- Read the contents of a folder (folder tech)
    - one row for each line
    - one row for each file

- It is better with a filename

```python
import os
# lecture des fichiers du répertoire
# defini un Dataframe
import pyspark.sql.functions
from pyspark.sql import functions as F

# read file 284.tex
df=spark.read.text("data/tech/284.txt")
print("*** Read File")
df.printSchema()
df.show(3,truncate=False)

# change the separator
df1=spark.read.text("data/tech/284.txt",lineSep="-")
print("*** Read File (-)")
df1.show(3,truncate=False)

# another choice
df2=spark.read.text("data/tech/284.txt",lineSep=".")
print("*** Read File (.)")
df2.show(2,truncate=False)

# read folder
df3=spark.read.text("data/tech/*")
print("*** Read Folder")
df3.show(5,truncate=False)

# one row - one file
df4=spark.read.text("data/tech/*", wholetext=True)
print("*** Read Folder (wholetext)")
df4.show(2,truncate=False)

# with the name of the file is usefull
df5 = df.withColumn('filename',F.input_file_name())
print("*** Read Folder with file name")
df5.printSchema()
df5.show(2, truncate = False)
```

```python
df = spark.read.csv("./data/zipcodes.csv")

df.printSchema()
df.show(truncate=False)
```

*.option* and *.options* in action

- define your delimiter

- read the header

- define your schema

```python
# define the delimiter
df3 = spark.read.option("delimiter",';').csv("data/zipcodes.csv")
print("*** delimiter")
df3.printSchema()
df3.show()

# define the delimiter
df3 = spark.read.options(delimiter=';').csv("data/zipcodes.csv")
print("*** delimiter")
df3.printSchema()
df3.show()


# use .options("header",True)
df2 = spark.read.options(header='True', delimiter=';') \
     .csv("data/zipcodes.csv")
print("*** delimiter + header")
df2.printSchema()
df2.show()
   
df6 = spark.read.options(header='True', inferSchema='True', delimiter=';') \
     .csv("data/zipcodes.csv")
print("*** delimiter + header + inferschema")
df6.printSchema()
df6.show() 

# check the file
# be carefull' 24' is not '24'
# define the schema
schema = StructType() \
      .add("id",IntegerType(),True) \
      .add("name",StringType(),True) \
      .add("complete_name",StringType(),True) \
      .add("year",IntegerType(),True) \
      .add("date",DateType(),True) \
      .add("date_hour",TimestampType(),True) \
      .add("boo_value",BooleanType(),True) \
      .add("dou_value",DoubleType(),True) \
      .add("int_Value",IntegerType(),True) \
      .add("boo_value2",BooleanType(),True) \
      .add("str_value",StringType(),True) \
      .add("missing_int_data1",IntegerType(),True) \
      .add("missing_str_data2",StringType(),True)

# define .option("dateFormat","???")
# define .option("TimeStampFormat","???")
df_with_schema = spark.read.format("csv") \
      .options(header=True,delimiter=";") \
      .schema(schema) \
      .option("dateFormat","MM/dd/yyyy")\
      .option("TimeStampFormat","MM/dd/yyyy HH:mm")\
      .load("data/zipcodes.csv")
print("*** delimiter + header + structtype")
df_with_schema.printSchema()

df_with_schema.show(5)
```

- Read a big CSV file

- define the schema

-

- select Date - Heure, date, year, month, Consommation brute électricité (MW) - RTE

```python
from pyspark.sql.functions import year, month, dayofmonth

# use .options("header",True)
# check the file
# be carefull' 24' is not '24'
# define the schema
schema = StructType() \
      .add("Date - Heure",TimestampType(),True) \
      .add("Date",DateType(),True) \
      .add("Heure",StringType(),True) \
      .add("Code INSEE région",StringType(),True) \
      .add("Région",StringType(),True) \
      .add("Consommation brute gaz (MW PCS 0°C) - GRTgaz",IntegerType(),True)\
      .add("Statut - GRTgaz",StringType(),True)\
      .add("Consommation brute gaz (MW PCS 0°C) - Teréga",IntegerType(),True)\
      .add("Statut - Teréga",StringType(),True) \
      .add("Consommation brute gaz totale (MW PCS 0°C)",IntegerType(),True)\
      .add("Consommation brute électricité (MW) - RTE",IntegerType(),True)\
      .add("Statut - RTE",StringType(),True) \
      .add("Consommation brute totale (MW)",IntegerType(),True)

df_big = spark.read.options(header='True', delimiter=';') \
     .schema(schema)\
     .csv("./data/consommation-quotidienne-brute.csv")
print("*** delimiter + header")
df_big.printSchema()
df_big.show(1,truncate=False)
df_big1 = df_big.select("Date - Heure","Date","Région",\
                        "Consommation brute électricité (MW) - RTE",\
                       year("Date").alias('year'),\
                       month("Date").alias('month'))
df_big1.printSchema()
df_big1.show(1, truncate = False)
```

https://spark.apache.org/docs/latest/api/python/reference/pyspark.sql/api/pyspark.sql.DataFrameReader.csv.html#pyspark.sql.DataFrameReader.csv

https://spark.apache.org/docs/latest/sql-data-sources-csv.html

```python
# define the delimiter
df = spark.read.json("data/zipcodes.json")
print("*** delimiter")
df.printSchema()
df.show()

# check the file
# be carefull' 24' is not '24'
# define the schema
schema = StructType() \
      .add("id",IntegerType(),True) \
      .add("name",StringType(),True) \
      .add("complete_name",StringType(),True) \
      .add("year",IntegerType(),True) \
      .add("date",DateType(),True) \
      .add("date_hour",TimestampType(),True) \
      .add("boo_value",BooleanType(),True) \
      .add("dou_value",DoubleType(),True) \
      .add("int_Value",IntegerType(),True) \
      .add("boo_value2",BooleanType(),True) \
      .add("str_value",StringType(),True) \
      .add("missing_int_data1",IntegerType(),True) \
      .add("missing_str_data2",StringType(),True)

# define .option("dateFormat","???")
# define .option("TimeStampFormat","???")
df_with_schema = spark.read.format("json") \
      .options(header=True,delimiter=";") \
      .schema(schema) \
      .load("data/zipcodes.json")
print("*** delimiter + header + structtype")
df_with_schema.printSchema()

df_with_schema.show()
```

root
 |-- data: struct (nullable = true)
 |    |-- emp_id: string (nullable = true)
 |    |-- awards: array (nullable = true)
 |    |    |-- element: struct (containsNull = true)

- define the right schema then read the test_emp.jason

show the columns emp_name and award_name


```python
schema = StructType() \
             .add("data", StructType() \
                   .add("emp_name", StringType(),True) \
                   .add("awards", ArrayType(StructType()                     
                       .add("award_type", StringType(), True) \
                       .add("award_name", StringType(), True)))
               , True)
 
df = spark.read.json(path= "data/test_emp.json", schema=schema, multiLine=True)
df.printSchema()
df.show(truncate=False)
df.select("data.emp_name").show()
df.select("data.awards.award_name").show(truncate=False)
```

### 4.1 Parquet


https://spark.apache.org/docs/latest/sql-data-sources-parquet.html

```python
peopleDF = spark.read.json("data/people.json")

# DataFrames can be saved as Parquet files, maintaining the schema information.
peopleDF.write.mode("overwrite").parquet("people.parquet")

# Read in the Parquet file created above.
# Parquet files are self-describing so the schema is preserved.
# The result of loading a parquet file is also a DataFrame.
parquetFile = spark.read.parquet("people.parquet")
parquetFile.show()
# Parquet files can also be used to create a temporary view and then used in SQL statements.
parquetFile.createOrReplaceTempView("parquetFile")
teenagers = spark.sql("SELECT * FROM parquetFile WHERE age >= 13 AND age <= 19")
teenagers.show()
# +------+
# |  name|
# +------+
# |Justin|
# +------+
```

```python
data = [('James','Smith','M',3000),
  ('Anna','Rose','F',4100),
  ('Robert','Williams','M',6200), 
]

columns = ["prenom","nom","genre","salaire"]
df = spark.createDataFrame(data=data, schema = columns)
df.show()


# partitionBy()

# Partitionner en fonction du genre
df.write.option("header",True)\
.partitionBy("genre")\
.mode("overwrite")\
.parquet("genretest.parquet")


# Partitionner en fonction du genre puis du salaire
df.write.option("header",True)\
.partitionBy("genre","salaire")\
.mode("overwrite")\
.parquet("genresalairetest.parquet")
                 

# A partir du répertoire précédent, lire les données de genre M et du salaire 4100
df1 = spark.read.parquet("genresalairetest.parquet/genre=M")
df1.show()
# A partir du répertoire précédent, lire les données de genre M
```

### 4.2 AVRO

https://sparkbyexamples.com/spark/read-write-avro-file-spark-dataframe/


```python
peopleDF = spark.read.json("data/people.json")

# DataFrames can be saved as Parquet files, maintaining the schema information.
peopleDF.write.format("avro").mode("overwrite").save("people.avro")
# Read in the Parquet file created above.
# Parquet files are self-describing so the schema is preserved.
# The result of loading a parquet file is also a DataFrame.
parquetFile = spark.read.format("avro").load("people.avro")
parquetFile.show()
# Parquet files can also be used to create a temporary view and then used in SQL statements.
parquetFile.createOrReplaceTempView("parquetFile")
teenagers = spark.sql("SELECT * FROM parquetFile WHERE age >= 13 AND age <= 19")
teenagers.show()
```
